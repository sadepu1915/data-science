```python
df=pd.read_csv('./data/EURUSD.csv', index_col='DATE', parse_dates=True)
```

```python
closing_price=df['CLOSE']
```

```python
closing_price
```

```python
mavg_50 = pd.rolling_mean(closing_price, 50)
mavg_100 = pd.rolling_mean(closing_price, 100)
mavg_200 = pd.rolling_mean(closing_price, 200)
```

```python
mavg_50[-10:]
```

```python
mavg_100[-10:]
```

```python
mavg_200[-10:]
```

```python
closing_price.plot(label='EURUSD')
mavg_50.plot(label='50 MVA')
mavg_100.plot(label='100 MVA')
mavg_200.plot(label='200 MVA')
plt.legend()
```
